```python
def parse_report(report):
    seen = False
    number = None
    obj_class = None
    containment_procedure = ""

    parts = report["text"].split("\n")
    parts.pop()
    for part in parts:
        if "Item #:" in part and seen == False:
            seen = True
            part = part.replace("–", "-")
            number = int(part.split("-")[1])
        elif "Object Class: " in part:
            obj_class = part.split("Object Class: ")[1]
        elif part.strip() != "" and len(part) > 15:
            containment_procedure += " " + part
    # print(number, obj_class, containment_procedure)
    if number == None or obj_class == None or containment_procedure == "":
        print(f"Part missing: number -> {number}, obj_class -> {obj_class}")

    return number, obj_class, containment_procedure
```

```python
def write_data(jobj, i):
    with open("./data/data.csv", "a") as f:
        writer = csv.writer(f, quoting=csv.QUOTE_MINIMAL)
        for index, row in jobj.iterrows():
            try:
                number, obj_class, containment_procedure = parse_report(row)
                writer.writerow([i, index, number, obj_class])
            except:
                print(i, row)
```

```python
for i in range(1, 9):
    obj = pd.read_json(path_or_buf=f"./data/data{i}.jsonl", lines=True)
    write_data(obj, i)
```
